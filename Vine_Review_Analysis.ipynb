```python
import pandas as pd
df = pd.read_csv("vine_table.csv", header='infer')
df.head()
```

```python
votes_df = df[df.total_votes >= 20]
votes_df.head()
```

```python
helpful_df = votes_df[votes_df.helpful_votes / votes_df.total_votes >= 0.5]
helpful_df.sample(10)
```

```python
yes_df = helpful_df[helpful_df.vine == "Y"]
yes_df.head()
```

```python
no_df = helpful_df[helpful_df.vine == "N"]
no_df.head()
```

```python
total_reviews_yes = yes_df["review_id"].count()
total_reviews_no = no_df["review_id"].count()
five_yes = yes_df[yes_df.star_rating == 5]["review_id"].count()
five_no = no_df[no_df.star_rating == 5]["review_id"].count()
pct_yes = round(five_yes * 100 / total_reviews_yes, 2)
pct_no = round(five_no * 100/ total_reviews_no, 2)
print(f"Vine: {five_yes} / {total_reviews_yes} = {pct_yes} \n\
Non-Vine: {five_no} / {total_reviews_no} = {pct_no}")
```

```python
Vine = [five_yes, total_reviews_yes, pct_yes]
NonVine = [five_no, total_reviews_no, pct_no]
data=[Vine, NonVine]

results = pd.DataFrame(data=data, index=["Vine", "Non-Vine"], columns=["Five-Star Reviews", "Total Reviews", "Five-Star Percentage of Total"])
results
```
